# ANALYZE PATHWAY DATA FROM IPA

```python
# Given the expoerted Networks we find all realtionships between them for VR
# ENSURE THAT THE FIRST LINE AFTER EXPORTING FROM IPA IS REMOVED!!!!! (not the header)


#list to store ALL "[from, to]" modelecules
fr_rel_to_list = []
#list to store "molecule: count"
dict_molecules = {}
    
for a in xrange(1,61): # range 1 to # of files + 1

    
    #read in the data file
#     df = pd.read_csv('test2.txt', sep='\t')
    name = 'exports/splits/vr/Network ' + str(a) + '.txt'
#     name = 'test' + str(a) + '.txt'
    df = pd.read_csv(name, sep='\t', names= ['From Molecule(s)', 'Relationship Type', 'To Molecule(s)', 'Catalyst(s)', 'Unnamed: 4'],
                    skiprows=1)
    
    from_column = df["From Molecule(s)"]
    to_column = df["To Molecule(s)"]
#     relation_column = df['Relationship Type']

    # we will first iterate through all the lines of the file
    for i in xrange(len(df)):
        from_molecule = from_column[i]
        to_molecule = to_column[i]
#         relationship = relation_column[i]
#       temp_list = [from_molecule, relationship, to_molecule]
        #create list to store "[from, to]"
        temp_list = [from_molecule, to_molecule]
        if from_molecule not in dict_molecules:
            dict_molecules[from_molecule] = 0
        if to_molecule not in dict_molecules:
            dict_molecules[to_molecule] = 0
        # it the list is new and the molecules are different we add them to the dictionary
        if temp_list not in fr_rel_to_list and from_molecule != to_molecule:
            fr_rel_to_list.append(temp_list)
            dict_molecules[from_molecule] += 1
            dict_molecules[to_molecule] += 1

with open('exports/results/vr_interations_count_incomplete.txt', 'w') as csvfile: # CHANGE THE NAME HEREEEEE!!!!!!!!!!!!!
    
    for key, value in dict_molecules.iteritems():
        string = str(key) + '\t' + str(value) + '\n'
        csvfile.write(string)
print "done"
```

```python
# plot the dict_molecules results

#REQUIREMENTS TO RUN: HAVE THE DICTIONARYs dict_molecules_nvr and dict_molecules

import collections

# gets only the counts from each variable
vals = dict_molecules.values()
counter=collections.Counter(vals)
y_vals = counter.values() # the counts
x_vals = counter.keys()  # the number of time a count appears
x_vals, y_vals = zip(*sorted(zip(x_vals, y_vals)))
x_vals, y_vals = (list(t) for t in zip(*sorted(zip(x_vals, y_vals))))
y_vals[:] = [x / float(len(dict_molecules)) for x in y_vals]

#do the above for nvr
vals_nvr = dict_molecules_nvr.values()
counter_nvr =collections.Counter(vals_nvr)
y_nvr_vals = counter_nvr.values() # the counts
x_nvr_vals = counter_nvr.keys()  # the number of time a count appears
# sort the values
x_nvr_vals, y_nvr_vals = zip(*sorted(zip(x_nvr_vals, y_nvr_vals)))
x_nvr_vals, y_nvr_vals = (list(t) for t in zip(*sorted(zip(x_nvr_vals, y_nvr_vals))))
y_nvr_vals[:] = [x / float(len(dict_molecules_nvr)) for x in y_nvr_vals]

fig = plt.figure(1)
p1 = plt.plot(x_vals, y_vals, 'r')
p2 = plt.plot(x_nvr_vals, y_nvr_vals, 'b')
plt.legend( (p1[0], p2[0]), ( 'VR', 'NVR' ) )
plt.xlabel('Number of Interactions')
plt.ylabel('Percentage %')
plt.title('Plot of Intercations for the NVR and VR')
plt.grid(True)
plt.show()

# fig.savefig('exports/results/plot_of_entire_interactions_in_NVR_and_VR_percentage.png')
```

```python
# Using dict_molecules, we count the genes (Unique Start Stop)  which are
# entirely in a non-variable region (Overlap type 5 or 7) then we compare to
# find the matching RefSeq to Molecule ID from IPA. then pull these our from 
# dict_molecules, output to a file

#REQUIREMENTS TO RUN: HAVE THE DICTIONARY dict_molecules

# file from IPA with eported molecule name
df_sym = pd.read_csv('exports/symbols/combined.txt', sep='\t')
df = pd.read_csv('../results_variable_regions_all.txt', sep='\t')

dict_of_mol = dict_molecules
# dict_of_mol = {'VAMP3':3,'ERRFI1':4,'NPPB':2,'RNU5E-1':1,'KIAA2013':5,'PADI1':2,'PADI3':5,'MIR3972':1,'PADI4':1,'MIR4695':1,'mir-1290':1}
# have a dictionary that stores only the molecules that have overlaptype 5 pr 7
dict_entire_vr = {}

overlap_type = df['Overlap_Type']
gene_id = df['Overlapping_GeneID']
chr_num = df['chr']
sym_list = list(df_sym['Symbol'])
refseq_list = list(df_sym['ID'])

def getsymbol(refseq_id_list):
    """
        Function which take the refseq id from a gene of overlap type 5/7
        and reutrns the IPA symbol
    """
    #if the first ID is not known by IPA cycle through other options
    for refseq_id in refseq_id_list:
        if refseq_id in refseq_list:
            symbol_idx = refseq_list.index(refseq_id)
            #if you can find symbol reutrn the symbol else reutrn Nan
            return sym_list[symbol_idx]
    return 'NaN'


for ot,j,gn in zip(overlap_type,chr_num,gene_id):
    if ot != '-':
        #strip the string to an array
        ot_stip = [x.strip() for x in ot.split(',')]
        gn_stip = [x.strip() for x in gn.split(',')]
        
        # Get the genes that overlap with overlap type 5 or 7
        for ots,gns in zip(ot_stip, gn_stip):
            if ots == '7' or ots == '5':
                gn_split_again = [x.strip() for x in gns.split('/')]
                # get the symbol
                sym = getsymbol(gn_split_again)
                #error check in case symbol is not in dataset
                all_keys = dict_of_mol.keys()
                if sym != 'NaN' and sym in all_keys:
                    dict_entire_vr[sym] = dict_of_mol[sym]

with open('exports/results/genes_entirely_in_vr_incomplete.txt', 'w') as csvfile: # CHANGE THE NAME HEREEEEE!!!!!!!!!!!!!
    
    for key, value in dict_entire_vr.iteritems():
        string = str(key) + '\t' + str(value) + '\n'
        csvfile.write(string)            
        
print "done"
```

# Using IPA Data to Remove Duplicates

```python
df = pd.read_csv('ipa_vr_export.txt', sep='\t')

notes = df['Notes']
reqseq_id = df['ID']

count = 0
no_dups = []
in_duplicate_seq = False

while count < len(notes):
    if notes[count] == 'D': # we only want the symbol 'D' since it represents duplicates
         if in_duplicate_seq == False: # only add the first ID no other duplicates
                no_dups.append(reqseq_id[count])
                in_duplicate_seq = True
                
    else: # it is not a dupliacte so just add to list
        in_duplicate_seq = False
        no_dups.append(reqseq_id[count])
    
    count += 1
        
no_dups
```

```python
df_t = pd.DataFrame(list(no_dups))
df_t.to_csv('ipa_no_dups_vr.csv', index=False, header=None)
```

# Generate Genes That are in VRs and NVR

```python
# Get all the unique geneIDs from  gene_start_stop_no_dups.csv and results_non_variable_regions_all.txt
import re
df1 = pd.read_csv('gene_start_stop_no_dups.csv', header=None, names=['chr', 'start', 'stop', 'gid'])
df2 = pd.read_csv('non_variable_regions_and_genes_update.txt', sep='\t')
df3 = pd.read_csv('variable_regions_and_genes_update.txt', sep='\t')

gid_vals = df1['gid']
all_genes = []
for gidv in gid_vals:
    if '/' in gidv:
        split_gid_vals = gidv.split('/')
        for split_gid in split_gid_vals:
            all_genes.append(split_gid)
    else:
        all_genes.append(gidv)

# remove duplicates
all_genes = list(set(all_genes))

#get the unique gids for non_variable_regions_and_genes.txt
gida = df2['Overlapping_GeneID']
genes_overlapping_nvr = []

for i in gida:
    if i != '-':
        # remove commas and backslashes
        split_gid =  re.findall(r"[\w']+", i)
        genes_overlapping_nvr.extend(split_gid)

# remove duplicates
genes_overlapping_nvr = list(set(genes_overlapping_nvr))


#get the unique gids for variable_regions_and_genes.txt
gida = df3['Overlapping_GeneID']
genes_overlapping_vr = []

for i in gida:
    if i != '-':
        # remove commas and backslashes
        split_gid =  re.findall(r"[\w']+", i)
        genes_overlapping_vr.extend(split_gid)

# remove duplicates
genes_overlapping_vr = list(set(genes_overlapping_vr))


# sanity checker i.e., genes in VR + NVR should be >= to the total (due to overlaps)
print len(all_genes)
print len(genes_overlapping_vr) + len(genes_overlapping_nvr)
# ensure that all the unique genes in nvr and vr are equal the total unqiue genes
all_g = genes_overlapping_vr
all_g.extend(genes_overlapping_nvr)
print len(set(all_g))
```

```python
df_t = pd.DataFrame(list(all_genes))
df_t.to_csv('all_genes.csv', index=False, header=None)

df_t = pd.DataFrame(genes_overlapping_nvr)
df_t.to_csv('genes_overlapping_nvr_update.csv', index=False, header=None)

df_t = pd.DataFrame(genes_overlapping_vr)
df_t.to_csv('genes_overlapping_VR_update.csv', index=False, header=None)
```

```python
# Total # of (Refseq ID)  overlapping a non-variable region by 50% reciprocally: 
# 2678  ( 2510 Autosomal, 155 X, 13 Y)

# Total # of (Refseq ID)  overlapping a non-variable region by less than 50% reciprocally: 
# 41983 (39410 Autosomal, 2361 X, 212 Y)

# Recipricol (>=50% and < 50%)

df = pd.read_csv('../results_non_variable_regions_all_reQ.txt', sep='\t')
df_gene = df['Percentage_of_Gene_Overlapping_NVR']
df_nvr = df['Percentage_of_NVR_Overlapping_Gene']
df_chr =df['chr']

all_greater_50 = 0 # all genes
auto_greater_50 = 0 # autosomal genes
x_greater_50 = 0 # X genes
y_greater_50 = 0 # Y genes

all_less_50 = 0 # all genes
auto_less_50 = 0 # autosomal genes
x_less_50 = 0 # X genes
y_less_50 = 0 # Y genes

for g, n, c in zip(df_gene, df_nvr, df_chr):
    if g != 0:
        if type(g) == type('str'):
            #strip the string to an array
            gen_strip = [x.strip() for x in g.split(',')]
            nvr_strip = [x.strip() for x in n.split(',')]
            
        # add to the set
        for gs, nvr in zip(gen_strip, nvr_strip):
            all_count += 1
            if float(nvr) >= 50 and float(gs) >= 50:
                all_greater_50 += 1
                if c != 'X' and c != 'Y':
                    auto_greater_50 += 1
                elif c == 'X':
                    x_greater_50 += 1
                else:
                    y_greater_50 += 1
            else:
                all_less_50 += 1
                if c != 'X' and c != 'Y':
                    auto_less_50 += 1
                elif c == 'X':
                    x_less_50 += 1
                else:
                    y_less_50 += 1
        
    
print all_greater_50
print auto_greater_50
print x_greater_50
print y_greater_50

print all_less_50
print auto_less_50
print x_less_50
print y_less_50
```

```python
# PLOT OF CHR # VS # OF GENES FOR REFSEQ
x1 = np.array([3517, 2283, 1993, 1371, 1696, 1868, 1706, 1315, 1412, 1447, 2091, 1772, 758, 1137, 1224, 1435, 1944, 589, 2187, 993, 511, 808, 1506, 197])
x2 = np.array([1100, 738, 623, 502, 525, 589, 350, 383, 571, 440, 633, 562, 279, 336, 315, 199, 505, 175, 374, 234, 139, 114, 568, 60] )

N =24 # number of bins
hist1 = [0] * (N)
hist2 = [0] * (N)
line_x = [0] * (N)
line_y = [0] * (N)

#--- create two histogram. Values of 1 go in Bin 0 ---
count = 0
for x, xx in zip(x1,x2):
    hist1[count] = x
    hist2[count] = xx-x
    line_x[count] = count+1
    line_y[count] = xx/float(x)
    count += 1

print np.mean(line_y)

plt.subplot(2, 1, 1)
width = 1
p1 = plt.bar( np.arange(0,N)+0.5, hist1, width, color='y' )
p2 = plt.bar( np.arange(0,N)+0.5, hist2, width, color='m', bottom=hist1 )
plt.legend( (p1[0], p2[0]), ( 'Genes in NVR', 'Genes' ) )
plt.xlabel( 'Chromosome' )
plt.ylabel( 'Number of Genes' )
plt.xticks( np.arange( 1,N+1 ) )

plt.subplot(2, 1, 2)
plt.plot(line_x, line_y, 'r.-')
plt.xlabel('Chromosome')
plt.ylabel('Percentage Overlap')
plt.xticks( np.arange( 1,N+1 ) )

plt.show()
```
